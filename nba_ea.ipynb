## Librerias Utilizadas 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Base de datos de NBA
https://www.kaggle.com/datasets/isaienkov/nba-top-50-alltime-scorers

```python
df = pd.read_csv("dataset/nba.csv", sep=",", 
dtype={'rank':'int64','player':'string','position':'string','total_points':'int64','total_games':'int64','points_per_game':'float64','field_goals':'int64', 'three_points_goals':'int64','free_shots':'int64','born':'string','active_player':'int64','country':'string'})
df.head(5)
```

### Missing Data

```python
plt.figure(figsize=(12,8)) 
sns.heatmap(df.isnull(),cmap='cool',cbar=False,yticklabels=False)
plt.title('Missing Data?',fontsize=20)
plt.show()
```

```python
df.plot(kind="scatter", x="total_games", y="total_points");
plt.show()
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(df, hue="country", height=8).map(plt.scatter, "total_games", "total_points").add_legend();
plt.show();
```
